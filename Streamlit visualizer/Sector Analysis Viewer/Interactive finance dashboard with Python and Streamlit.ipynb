```python
import sqlalchemy
import yfinance as yf
import pandas as pd
```

```python
tickers = pd.read_html(
    'https://en.wikipedia.org/wiki/Dow_Jones_Industrial_Average')[1]
```

```python
tickers = tickers.Symbol.to_list()
```

```python
df = pd.DataFrame()
```

```python
for ticker in tickers:
    var = yf.Ticker(ticker).info
    frame = pd.DataFrame([var])
    df = df.append(frame)
```

```python
engine = sqlalchemy.create_engine('sqlite:///Fundamentals.db')
```

```python
subdf = df[['symbol', 'dividendYield', 'sector', 'shortRatio', 'forwardPE', 'trailingPE']]
```

```python
subdf
```

```python
subdf.to_sql('Fundamentalstable', engine,index=False)
```

```python
pd.read_sql('Fundamentalstable', engine, index_col = ['symbol'])
```
